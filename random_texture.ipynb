### Perlin-like Color Field

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter

# np.random.seed(42)
data = np.random.rand(200, 200)
blurred = gaussian_filter(data, sigma=10)

plt.imshow(blurred, cmap='Blues')
plt.title("Perlin-like Color Field")
plt.axis('off')
plt.show()
```

### Sinusoidal Gradient Patch

```python
x = np.linspace(0, 3 * np.pi, 400)
y = np.linspace(0, 3 * np.pi, 400)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) + np.cos(Y)

plt.imshow(Z, cmap='Spectral')
plt.title("Sinusoidal Gradient Patch")
plt.axis('off')
plt.show()
```

### RGB Pattern Blend

```python
width, height = 256,256
R = np.random.rand(height, width)
G = np.cos(np.linspace(0, 10, width))[None, :]
B = np.sin(np.linspace(0, 10, height))[:, None]

img = np.stack([R, G.repeat(height, 0), B.repeat(width, 1)], axis=2)
plt.imshow(img)
plt.axis('off')
plt.title("RGB Pattern Blend")
plt.show()
```

### Mandelbrot Fractal Patch

```python
def mandelbrot(c, max_iter):
    z = 0
    for i in range(max_iter):
        z = z*z + c
        if abs(z) > 2:
            return i
    return max_iter

w, h = 300, 300
img = np.zeros((h, w))
for x in range(w):
    for y in range(h):
        c = complex((x - w/2) / 100, (y - h/2) / 100)
        img[y, x] = mandelbrot(c, 50)

plt.imshow(img, cmap='twilight_shifted')
plt.title("Mandelbrot Fractal Patch")
plt.axis('off')
plt.show()
```

## Math Art Generator - 4 Styles in One Script

```python
# Math Art Generator - 4 Styles in One Script

import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter

# Set canvas size
width, height = 300, 300

# ---------- Style 1: Perlin-like Noise (Blurry Random Field) ----------
def generate_perlin_art():
    data = np.random.rand(height, width)
    blurred = gaussian_filter(data, sigma=10)
    plt.imshow(blurred, cmap='plasma')
    plt.axis('off')
    plt.title("Perlin-like Color Field")
    plt.show()

# ---------- Style 2: Sinusoidal Gradient Pattern ----------
def generate_sinusoidal_art():
    x = np.linspace(0, 3 * np.pi, width)
    y = np.linspace(0, 3 * np.pi, height)
    X, Y = np.meshgrid(x, y)
    Z = np.sin(X) + np.cos(Y)
    plt.imshow(Z, cmap='Spectral')
    plt.axis('off')
    plt.title("Sinusoidal Gradient Patch")
    plt.show()

# ---------- Style 3: RGB Function Blending ----------
def generate_rgb_patch_art():
    R = np.random.rand(height, width)
    G = np.sin(np.linspace(0, 10, width))[None, :]
    B = np.cos(np.linspace(0, 10, height))[:, None]
    img = np.stack([R, G.repeat(height, 0), B.repeat(width, 1)], axis=2)
    plt.imshow(img)
    plt.axis('off')
    plt.title("RGB Pattern Blend")
    plt.show()

# ---------- Style 4: Fractal (Mandelbrot Set) ----------
def mandelbrot(c, max_iter):
    z = 0
    for i in range(max_iter):
        z = z*z + c
        if abs(z) > 2:
            return i
    return max_iter

def generate_mandelbrot_art():
    max_iter = 50
    img = np.zeros((height, width))
    for x in range(width):
        for y in range(height):
            c = complex((x - width/2) / 100, (y - height/2) / 100)
            img[y, x] = mandelbrot(c, max_iter)
    plt.imshow(img, cmap='twilight_shifted')
    plt.axis('off')

    plt.show()

# ---------- Run All Styles ----------
if __name__ == "__main__":
    generate_perlin_art()
    generate_sinusoidal_art()
    generate_rgb_patch_art()
    generate_mandelbrot_art()
```

## Soft Blurry Paint Strokes

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter

np.random.seed(0)
noise = np.random.rand(100,100)
blurred = gaussian_filter(noise, sigma=12)

plt.imshow(blurred, cmap='plasma')  # Try 'magma', 'cividis', 'twilight'
plt.title("Soft Blurry Paint Strokes")
plt.axis('off')
plt.show()
```

## Sharp Digital Blocks

```python
x = np.linspace(0, 10, 300)
y = np.linspace(0, 10, 300)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

Z_quantized = np.floor(Z * 5) / 5  # Increase 5 → more steps, lower → blockier

plt.imshow(Z_quantized, cmap='viridis')
plt.axis('off')
plt.title('Sharp Digital Blocks')
plt.show()
```

## Gradient vibes

```python
np.random.seed(1)
noise1 = gaussian_filter(np.random.rand(300, 300), sigma=50)
noise2 = gaussian_filter(np.random.rand(300, 300), sigma=50)
combined = (noise1 + noise2) / 2

plt.imshow(combined, cmap='plasma')  # Or 'inferno', 'coolwarm', 'gist_stern'
plt.axis('off')
plt.title("Gradient vibes")
plt.show()
```

## Uzumaki

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 4 * np.pi, 10000)
r = theta**0.8  # Spiral growth

x = r * np.cos(theta + 0.3 * np.sin(3*theta))
y = r * np.sin(theta + 0.3 * np.sin(3*theta))
colors = np.sin(theta * 5)  # Color variation

plt.figure(figsize=(2, 2))
plt.scatter(x, y, c=colors, cmap='twilight_shifted', s=0.5)
plt.title('Uzumaki (Swirl)')
plt.axis('off')
plt.show()
```

## Starfield with Glow

```python
from scipy.ndimage import gaussian_filter

size = 512
stars = np.zeros((size, size))

# Add bright points
for _ in range(500):
    x, y = np.random.randint(0, size, 2)
    stars[y, x] = np.random.rand() * 5

# Create glow with blur
glow = gaussian_filter(stars, sigma=1.5)

plt.imshow(glow, cmap='magma')
plt.axis('off')
plt.title('Starfield with Glow')
plt.show()
```

## Circular ripples on water

```python
x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y)

Z = np.sin(X**2 + Y**2) + np.cos(3 * np.sqrt(X**2 + Y**2))

plt.imshow(Z, cmap='Blues', alpha=0.8)
plt.axis('off')
plt.title('Circular ripples on water')
plt.show()
```
